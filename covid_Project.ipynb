```python
def weibull(t,scale=120,shape=3.67):
  exponent = math.pow( t/scale , shape )
  return math.exp( -1*exponent )
```

```python
## parameters is of order
## beta, St, Et, It, Rt, CIRt, M , dVt, dWt, N

def Next_Instant_Values(parameters):
  beta = parameters[0]    ## List of Beta values for all the units - For all the List the size should be 38 X 1
  St = parameters[1]      ## List for all units - Susceptible at time t
  Et = parameters[2]      ## List for all units - Exposed at time t
  It = parameters[3]      ## List for all units - Infection at time t
  Rt = parameters[4]      ## List for all units - Recovered at time t
  CIRt = parameters[5]    ## List for all units - CIR rate at time t. We can take it as a constant - also we can vary it using number of tests
  M = parameters[6]       ## This is the mobility Matrix - This should be of 38 X 38
  dVt = parameters[7]     ## List for all units - Number of vaccinations done on the current date
  dWt = parameters[8]     ## List for all units - Number of Waned population at current time
  N = parameters[9]       ## List for all units - Number of peoples 
  Vt = parameters[10]     ## List for all units - Vaccination % at time t -- Currently we are not using it

  epsilon = 0.66 ## Vaccination Efficiency
  alpha = 1 / 5.8 ## Rate from Expose to Infection
  gamma = 1 / 5 ## Rate from Infectioin to Recovered

  Snext = []
  Enext = []
  Inext = []
  Rnext = []
  Vnext = []

  for i in range(38):
    ## This is for this unit
    d1 = alpha * Et[i]
    d2 = gamma * It[i]
    d3 = epsilon * dVt[i]
    d4 = 0

    for j in range(38):
      d4 += M[i][j] * It[j] / N[j]
    
    d4 *= St[i] * beta[i]

    ## These are the differences

    snext = int( St[i] - d4 - d3 + dWt[i] )
    enext = int( Et[i] + d4  - d1 )
    inext = int( It[i] + d1 - d2 )
    rnext = int( Rt[i] + d2 - dWt[i] )
    vnext = int( Vt[i] + d3 )

    Snext.append(snext)
    Enext.append(enext)
    Inext.append(inext)
    Rnext.append(rnext)
    Vnext.append(vnext)
  
  return ( Snext , Enext , Inext , Rnext , Vnext )
```

```python
def Loss1(parameters):
  beta = parameters[0]    ## List of Beta values for all the units - For all the List the size should be 38 X 1
  S0 = parameters[1]      ## List for all units - Susceptible at time t
  E0 = parameters[2]      ## List for all units - Exposed at time t
  I0 = parameters[3]      ## List for all units - Infection at time t
  R0 = parameters[4]      ## List for all units - Recovered at time t
  CIR0 = parameters[5]    ## List for all units - CIR rate at time t. We can take it as a constant - also we can vary it using number of tests
  M = parameters[6]       ## This is the mobility Matrix - This should be of 38 X 38
  N = parameters[7]       ## List for all units - Number of peoples 
  V0 = parameters[8]      ## List for all units - Vaccination % at time t -- Currently we are not using it
  scale = parameters[9]
  shape = parameters[10]

  epsilon = 0.66 ## Vaccination Efficiency
  alpha = 1 / 5.8 ## Rate from Expose to Infection
  gamma = 1 / 5 ## Rate from Infectioin to Recovered

  S21days = [S0]
  E21days = [E0]
  I21days = [I0]
  R21days = [R0]
  V21days = [V0]
  CIR21days = [CIR0]

  delta_R = [R0]

  ## dVt and dWt we will get from the timeseries data and function respectively

  days = 21 ## From 11Oct - 31Oct and 1 Nov

  for i in range(21):
    
    '''
      Compute CIR if it is dynamic. Otherwise keep it constant
    '''

    ## print("Day = " , i)

    CIRcur = CIR0
    
    Scur = S21days[i]
    Ecur = E21days[i]
    Icur = I21days[i]
    Rcur = R21days[i]
    Vcur = V21days[i]

    delta_V = np.zeros(38)
    delta_W = []

    for j in range(38):
      k = i
      wane_population = 0
      while (k >= 0):
        wane_population += delta_R[k][j]*(1 - weibull(i - k + 1 , scale , shape))
        k-=1
      delta_W.append(wane_population)


    Snext, Enext, Inext, Rnext, Vnext = Next_Instant_Values( (beta, Scur, Ecur, Icur, Rcur, CIRcur, M, delta_V, delta_W, N, Vcur) )

    S21days.append(Snext)
    E21days.append(Enext)
    I21days.append(Inext)
    R21days.append(Rnext)
    V21days.append(Vnext)
    CIR21days.append(CIRcur)

    delta_R_cur = []

    for j in range(38):
      delta_R_cur.append(Rnext[j] - Rcur[j])

    delta_R.append(delta_R_cur) ## These are the people who went to recovered this day
  
  ## After calculating the 21days values we have to find the loss
  ## Reported_Cases_On_1_Nov = [] ## Find out this from the time series data
  Predicted_Case_On_1_Nov = []

  '''
    For finding the Predicted cases on 1Nov we have to take running avg of past 7 days alpha * e
  '''
  for i in range(38):
    val = 0
    for j in range(7):
      val += alpha * (E21days[-1*j - 2][i] / CIR21days[-1*j - 2][i])
    val /= 7
    Predicted_Case_On_1_Nov.append(val)
  
  ## Now compute the loss
  loss = 0.0
  
  for i in range(38):
    diff = abs( Reported_Cases_On_1_Nov[i] - Predicted_Case_On_1_Nov[i] ) / Reported_Cases_On_1_Nov[i]
    loss = max(loss , diff)
    
  
  print(loss)
  print(Reported_Cases_On_1_Nov)
  print(Predicted_Case_On_1_Nov)

  #assert(1 == 2)

  return loss
```

```python
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                           ## It should take the initial parameters and return the converged parameters

def gradient_descent1(parameters):
  beta = parameters[0]
  scale = parameters[1] ## This is the weibull scale parameter
  shape = parameters[2] ## This is the weibull shape parameter

  ## Below values are to be gotten from the seroprevalence data and timeseries data
  '''
  S0 = []
  E0 = []
  I0 = []
  R0 = []
  V0 = []
  CIR0 = []
  N = []
  '''
  M = np.identity(38)   ## This is because for this task the mobility is Identity

  ## We have to change the beta values as 0.01, scale values as 1 and shape values as 0.01

  iter = 0

  min_loss = 100000000.0
  min_parameters = (beta , scale , shape)


  while (True):

    print(" _______________ iteration " , iter , " is starting _____________ ")
    iter+=1

    print(min_parameters)

    iter_parameters = min_parameters

    changed = False

    changes = [0.2 , 1, 0.01] ## First one is for the beta, second for the scale and last for the shape

    ## First lets change the beta values
    for i in range(38):
      ## print(i)
      beta_cur_pos = iter_parameters[0]
      beta_cur_neg = iter_parameters[0]
      beta_cur_pos[i] += changes[0]
      beta_cur_neg[i] -= changes[0]

      ## print("I am here")
      Beta_pos_val = Loss1( (beta_cur_pos, S0, E0, I0, R0, CIR0, M, N, V0, iter_parameters[1], iter_parameters[2]) )
      ## print("I am here")
      Beta_neg_val = Loss1( (beta_cur_neg, S0, E0, I0, R0, CIR0, M, N, V0, iter_parameters[1], iter_parameters[2]) )
      ## print("I am here")

      if (Beta_pos_val < min_loss):
        min_loss = Beta_pos_val
        min_parameters = (beta_cur_pos , scale , shape)
        changed = True
      
      if (Beta_neg_val < min_loss):
        min_loss = Beta_neg_val
        min_parameters = (beta_cur_neg , scale , shape)
        changed = True
    
    ## Now lets change the scale
    Scale_pos_val = Loss1( (iter_parameters[0], S0, E0, I0, R0, CIR0, M, N, V0, iter_parameters[1] + changes[1] , iter_parameters[2]) )
    Scale_neg_val = Loss1( (iter_parameters[0], S0, E0, I0, R0, CIR0, M, N, V0, iter_parameters[1] - changes[1] , iter_parameters[2]) )

    if (Scale_pos_val < min_loss):
      min_loss = Scale_pos_val
      min_parameters = ( iter_parameters[0] , iter_parameters[1] + changes[1] , iter_parameters[2] )
      changed = True
    
    if (Scale_neg_val < min_loss):
      min_loss = Scale_neg_val
      min_parameters = ( iter_parameters[0] , iter_parameters[1] - changes[1] , iter_parameters[2] )
      changed = True
    
    ## Now lets change the shape
    Shape_pos_val = Loss1( (iter_parameters[0], S0, E0, I0, R0, CIR0, M, N, V0, iter_parameters[1] , iter_parameters[2] + changes[2] ) )
    Shape_neg_val = Loss1( (iter_parameters[0], S0, E0, I0, R0, CIR0, M, N, V0, iter_parameters[1] , iter_parameters[2] - changes[2] ) )

    if (Shape_pos_val < min_loss):
      min_loss = Shape_pos_val
      min_parameters = ( iter_parameters[0] , iter_parameters[1] , iter_parameters[2] + changes[2]  )
      changed = True
    
    if (Shape_neg_val < min_loss):
      min_loss = Shape_neg_val
      min_parameters = ( iter_parameters[0] , iter_parameters[1] , iter_parameters[2] - changes[2]  )
      changed = True
    
    print("Loss in this iteration = " , min_loss)

    ## assert(1 == 2)

    if(changed == False):
      print("Gradient is not going down so exiting")
      break
```

```python
def dummyloss(x): 
  M = np.identity(38)
  return Loss1( (x[:38], S0, E0, I0, R0, CIR0, M, N, V0, x[38], x[39]) )
```
